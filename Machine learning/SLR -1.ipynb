```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge


from sklearn.metrics import r2_score
```

```python
data=sns.load_dataset('penguins')
```

```python
data.dropna(inplace=True)
```

```python
sns.scatterplot(x=data['flipper_length_mm'],y=data['body_mass_g'])
```

#### this two variable has linear relationship , Hence we can create a linear regression for this dataset

```python
x = pd.DataFrame(data['flipper_length_mm'])
y =pd.DataFrame( data['body_mass_g'])
```

### Feature Scaling 

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
y=scaler.fit_transform(y)
```

```python
x.shape , y.shape
```

### train test split of the datapoints

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=43)
```

### Linear regresssion

```python
regressor=LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
ypred=regressor.predict(xtest)
```

### Accuracy

```python
print( f"model accuracy : {round(r2_score(ytest,ypred),2)*100} %")
```
