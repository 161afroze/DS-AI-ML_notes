1) SUPERVISED LEARNING  


2) UNSUPERVISED LEARNING


3) REINFORCEMENT LEARNING



SUPERVISED LEARNING : supervised learning is divided into two types and they are  

1) REGRESSSION

2) CLASSIFICATION

 ### REGRESSION :

In this type of learning , input and ouput variables are already given to the model , so it it learns how an ouput is been formed from an input .

It tries to create a mapping from input to ouput and learn the pattern.

It tries to understand the realationship between the input and output and when a new input is given to it , it responds with the ouput according to the pattern found in given data

Regression is about predictin the continous values

example : predicting the stock price , predicting the house price, predicting the age
