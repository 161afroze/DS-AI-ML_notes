### QUARTILE ,  QUANTILE and PERCENTIILE

----------------------------------------------------------------------------------------------------------

Quartiles are a set of three values that divide ordered data sets into four equal parts. 

Quantiles are units of equal, adjacent quantities in a distribution.

----------------------------------------------------------------------------------------------------------

we have 3 Quartiles 

Q1 aka lower quartile (25th) percentile or qunatile 

Q2 aka median quartile (50th) percentile or qunatile

Q3 aka upper quartile (75th) percentile or qunatile



at the  point Q1 -25% of the values of the data will fall below this point 

at the point Q2 - 50% of values of the data will fall below this point 

at the point Q3 - 75% of values of the data will fall below this point 

```python
import numpy as np

#consider this are the marks of the student
data = np.array([15, 20, 35, 67, 71, 78, 82, 86, 90, 96, 98])
```

### Now the principal of a school asked his faculty "what it the 50th percentile marks in the classroom?"

```python
print(f"50% percent of student score more than {np.percentile(data,50) } marks")
```

```python
print(f"50% percent of student score more than {np.quantile(data,0.5) } marks")
```



percentiles are expressed in 1 to 100 ->(25% , 50% , 75% 100% )

quantiles are expressed in 0 to 1 -> (0 , 0.25 , 0.50 , 0.75 .1)

so 

--> 0.25 quantile == 25% percentile 

--> 0.50 quantile == 50% percentile

### "if a student had scored  89 , in which quadrant he will be in ?"

```python
print(f" quartile one : {np.quantile(data,0.25)}")
print(f" quartile two : {np.quantile(data,0.5)}")
print(f" quartile three : {np.quantile(data,0.75)}")
```

### answer : if a students scores 89 , he will fall into the upper quadrant or Q3
