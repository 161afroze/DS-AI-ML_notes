### Feature Scaling 

It is the technique where we convert the different numerical features of the dataset into a similar scale.This is done so that no other feature may dominated on other feature and significantly distruring the learning algorithm 

all the features are brought down to same unit of measurement 

## Various technique in Feature Scaling 

1) -> Standardization 
2) -> Normalization 
3) -> unit vector  


### Standardization technique

In this technique we try to convert it into the values where its around the its(data's) mean and with 1 standard devaiton 

mostly when the data is normally distriubuted then the values ranges form -3 to 3 

this standardization is done through the means of z score formula , and sometimes this technique is also called as zscore normalization 

#### zscore formula : 


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('tips')
data=data[['total_bill','tip']]
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

data_scaled=pd.DataFrame(scaler.fit_transform(data))
data_scaled
```

```python
data_scaled.describe()
```

### Normalization technique

normalization technique is another feature scaling techinque where the data is converted into range of  0 and 1 

This technique is also called as min max scaler 

#### maxmin Formula:

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler=MinMaxScaler()

data_scaled_minmax=pd.DataFrame(minmax_scaler.fit_transform(data))
data_scaled_minmax
```

```python
data_scaled_minmax.describe()
```
