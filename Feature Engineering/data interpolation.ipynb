## what is data interpolation

interpolation can be said as creating  new data points with in a range 

## why do we need interpolation    

### consider we have data where we have pirce of the gold on few days and now you want to get the price of the gold for some specific day , here through the concept of interpolation we can find out the price of the gold 

```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
days=[2,5,7]
gold_price=[88,76,77]

gold_price_interpreter=interp1d(x=days,y=gold_price)

price_on_Day=4
```

### what is the price on the 4th day of the link

```python
price=gold_price_interpreter(price_on_Day)
price
```

```python
print("--------------------------------------------------")
print(f"so the price is RS {price} on the {price_on_Day} ")
print("--------------------------------------------------")
```

### hence through the data interpolation method we have estimated  a new data point with respective to the data points we had This was the example of linear interpolation , as we had the linear data , when we have cubic or polynomial data then we need to use the cubic interpolation or polynoimal interpolation accordingly
