### what is range and IQR

range = highest_value-lowest_Value

IQR =Q3-Q1 (IQR stands for inter quartile range)

 


### how to calculate the Q1 and Q3 ? 

```python
import math
mylist=[1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,29]

Q1=math.ceil((25/100)*(len(mylist)))
Q3=math.ceil((75/100)*(len(mylist)))

print(f" Q1 : {Q1} and Q3 : {Q3}")
print("these are the index value and they are 1 based indexing")
```

so here the 5th index( 1 based) is the Q1 25th percentile 

and here the 15th index( 1 based ) is the Q3 75th percentile

## np.quantle() direct method , which will directly find the value from my the list

```python

import numpy as np
values=np.quantile(mylist,[0.25,0.75])
Q1value=int(values[0])
Q3value=int(values[1])
print(Q1value,Q3value)
```

by manually calculating we found the index (noticeable point)

but using the np.quantile method , it doest  not give you the index value it gives you result of the quantile

```python
import seaborn as sns
import numpy as np
data_points=[4,6,8,12,13,13,15,16,17,18,19,20,21,65,66,89,98,]
sns.boxplot(data_points)

lower_fence,Q1,median,Q3,upper_fence=np.quantile(data_points,[0,0.25,0.5,0.75,1])
print(f"lower_fence : {lower_fence} \n"+
      f"Q1 quadrant : {Q1} \n" +
      f"median : {median} \n"+
      f"Q3 quadrant : {Q3} \n"+
      f"maximum : {upper_fence}")
```
